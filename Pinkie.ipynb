```python
def random_quad(width, height, maxxangle, maxyangle, maxzangle):
    distfactor = 3.0
    distfactor2 = 1.0
    
    rotVect = np.zeros((3, 1))
    rotVect[0,0] = maxxangle * (2.0 * rand() - 1.0)
    rotVect[1,0] = (maxyangle - abs(rotVect[0,0])) * (2.0 * rand() - 1.0)
    rotVect[2,0] = maxzangle * (2.0 * rand() - 1.0)
    
    d = (distfactor + distfactor2 * (2.0 * rand() - 1.0)) * width

    rotMat, _ = cv2.Rodrigues(rotVect)
    
    
    quad = np.zeros((4, 2))
    
    halfw = 0.5 * width
    halfh = 0.5 * height
    
    quad[0,0] = -halfw;
    quad[0,1] = -halfh;
    quad[1,0] = halfw;
    quad[1,1] = -halfh;
    quad[2,0] = halfw;
    quad[2,1] = halfh;
    quad[3,0] = -halfw;
    quad[3,1] = halfh;
    
    quad[0,0] = -width / 2.0
    quad[0,1] = -height / 2.0
    
    for i in range(4):
        rotVect[0, 0] = quad[i, 0]
        rotVect[1, 0] = quad[i, 0]
        rotVect[2, 0] = quad[1, 0]
        vect = np.asmatrix(rotMat) * np.asmatrix(rotVect)
        quad[i, 0] = vect[0,0] * d / (d + vect[2, 0]) + halfw
        quad[i, 1] = vect[1,0] * d / (d + vect[2, 0]) + halfh
    
    return quad

```
